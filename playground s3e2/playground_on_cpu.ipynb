Stroke prediction task

```python
import numpy as np
import pandas as pd
import torch
from sklearn.preprocessing import OneHotEncoder
```

```python
test_dataset = pd.read_csv('test.csv')
test_dataset.head()
```

```python
train_dataset = pd.read_csv('train.csv')
train_dataset.head()
```

```python
X_train = train_dataset.drop('id', axis=1).drop('stroke', axis=1)
y_train = train_dataset['stroke']

X_test = test_dataset.drop('id', axis=1)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_train.info()
```

```python
to_encode = ['gender', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
```

```python
ohe = OneHotEncoder()
ohe.fit(X_train[to_encode])
new_cols = pd.DataFrame.sparse.from_spmatrix(ohe.transform(X_train[to_encode]))
X_train = X_train.drop(to_encode, axis=1)
X_train = pd.concat([X_train, new_cols], axis=1)

new_cols = pd.DataFrame.sparse.from_spmatrix(ohe.transform(X_test[to_encode]))
X_test = X_test.drop(to_encode, axis=1)
X_test = pd.concat([X_test, new_cols], axis=1)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_train = X_train.values
Y_train = y_train.values
X_test = X_test.values
```

Data manipulations are over

```python
#defining dataset class
from torch.utils.data import Dataset, DataLoader

class dataset(Dataset):
    def __init__(self,x,y):
        self.x = torch.tensor(x,dtype=torch.float32)
        self.y = torch.tensor(y,dtype=torch.long)
        self.length = self.x.shape[0]
 
    def __getitem__(self,idx):
        return self.x[idx],self.y[idx]
    def __len__(self):
        return self.length
trainset = dataset(X_train, Y_train)
#DataLoader
label_uniques, counts = np.unique(Y_train, return_counts=True)
class_weights = [sum(counts) / c for c in counts]
example_weights = [class_weights[e] for e in Y_train]
sampler = torch.utils.data.WeightedRandomSampler(example_weights, len(Y_train))


trainloader = DataLoader(trainset,batch_size=1024,shuffle=False)
```

```python
from torch import nn
from torch.nn import functional as F
```

```python
class Net(nn.Module):
    def __init__(self,input_shape):
        super(Net,self).__init__()
        self.fc1 = nn.Linear(input_shape,32)
        self.fc2 = nn.Linear(32,64)
        self.fc3 = nn.Linear(64,2)
    def forward(self,x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.sigmoid(self.fc3(x))
        return x
```

```python
#hyper parameters
learning_rate = 0.01
epochs = 100
# Model , Optimizer, Loss
model = Net(input_shape=X_train.shape[1])
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)
loss_fn = nn.NLLLoss(weight=torch.tensor(class_weights,dtype=torch.float32))
```

```python
from tqdm import tqdm
from sklearn.metrics import balanced_accuracy_score
losses = []
accur = []
for i in tqdm(range(epochs)):
    for j, (x_train, y_train) in enumerate(trainloader):
    
    #calculate output
        output = model(x_train)
 
    #calculate loss
        loss = loss_fn(output,y_train)
 
    #accuracy
        predicted = model(torch.tensor(X_train,dtype=torch.float32))
        acc = (np.argmax(predicted.detach().cpu().numpy(), axis=1) == Y_train).mean() 
        bacc = balanced_accuracy_score(Y_train, np.argmax(predicted.detach().cpu().numpy(), axis=1))
    #backprop
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    if i%10 == 0:
        losses.append(loss)
        accur.append(acc)
        print("\n epoch {}\tloss : {}\t balanced accuracy : {}\t accuracy: {}".format(i,loss,bacc,acc))
```

```python
prediction = model(torch.tensor(X_test,dtype=torch.float32))
prediction = prediction.reshape(-1).detach().numpy()
```

```python
prediction=pd.DataFrame(prediction, range(15304, 15304 + len(prediction)), ['stroke'])
prediction.head()
```
